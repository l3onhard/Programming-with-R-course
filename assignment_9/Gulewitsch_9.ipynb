```r
# Aufgabe 23

install.packages("mlbench")
library(mlbench)
data(PimaIndiansDiabetes)
diabetes.df <- PimaIndiansDiabetes
```

```r
## a)

?PimaIndiansDiabetes
str(diabetes.df)
summary(diabetes.df)
```

```r
# pregnant: Number of times pregnant; numeric
# glucose: Plasma glucose concentration (glucose tolerance test); numeric
# pressure: Diastolic blood pressure (in mm Hg); numeric
# triceps: Triceps skin fold thickness (in mm); numeric
# insulin: 2-Hour serum insulin (in mu U/ml); numeric
# mass: Body mass index (in weight in kg/(height in m)\^2b); numeric
# pedigree: Diabetes pedigree function; numeric
# age: Age (years); numeric
# diabetes: Class variable (test for diabetes); factor
```

```r
## b

diabetes.plotTitles <- c("Number of times pregnant", "Plasma glucose concentration",
                         "Diastolic blood pressure\n (in mm Hg)", "Triceps skin fold thickness\n (in mm)",
                         "2-Hour serum insulin (in mu U/ml)", "Body mass index\n (in weight in kg/(height in m)^2)",
                         "Diabetes pedigree function", "Age (in years)", "Diabetes diagnosis")
levels(diabetes.df$diabetes) <- c("negative", "positive")

for(index in seq_along(diabetes.df)){
    
    variable <- diabetes.df[[index]]
    title <- diabetes.plotTitles[index]
    
    if(class(variable) == "numeric"){
        boxplot(variable, main = title)
    }
 
    if(class(variable) == "factor"){
        barplot(table(variable), main = title, ylab = "Absolute Frequency")
    }
}
```

```r
## c)

?PimaIndiansDiabetes
```

```r
dirtyVariables <- c("glucose", "pressure", "triceps", "insulin", "mass")

for(variable in dirtyVariables){
    diabetes.df[variable][diabetes.df[variable] == 0] <- NA
}
```

```r
## d)

logVariables <- c("pedigree", "insulin")

for(variable in logVariables){
    hist(diabetes.df[[variable]])
    hist(log(diabetes.df[[variable]]))
}
```

```r
# The standard histograms of both variables ("pedigree" and "insulin") are right-skewed.
# When we use a logarithmic scale we get histograms that show distributions that appear normally distributed. 
```

```r
## e)

compare.df <- diabetes.df[- which(names(diabetes.df) == "diabetes")]
compare.plotTitles <- c("Number of times pregnant", "Plasma glucose concentration", 
                        "Diastolic blood pressure (in mm Hg)", "Triceps skin fold thickness (in mm)", 
                        "2-Hour serum insulin (in mu U/ml)", "Body mass index\n (in weight in kg/(height in m)^2)", 
                        "Diabetes pedigree function", "Age (in years)")

for(index in seq_along(compare.df)){
    variable <- compare.df[[index]]
    title <- paste(compare.plotTitles[index], "\nby diabetes diagnosis")
    ylab <- compare.plotTitles[index]
    par(mar = c(5, 5, 4, 2) + 0.1)
    boxplot(variable ~ diabetes.df$diabetes, main = title, 
            xlab = "Diabetes diagnosis", ylab = ylab)
}
```

```r
## f)

diabetes.df$pregnant <- cut(diabetes.df$pregnant, 
                            breaks = c(0, 1, 2, 3, max(diabetes.df$pregnant) + 1),
                            labels = c("=0", "=1", "=2", ">=3"),
                            ordered_results = TRUE, 
                            right = FALSE)
```

```r
cat("\nAbsulute Frequencies:")
(absoluteFrequencies <- table(diabetes.df$pregnant))
cat("\nRelative Frequencies:")
(relativeFrequencies <- absoluteFrequencies / sum(absoluteFrequencies))
```

```r
## g)

byPregnant.df <- diabetes.df[names(diabetes.df) %in%  c("pressure", "insulin", "mass")] 
```

```r
byPregnant.summary <- summary(byPregnant.df)
```

```r
byPregnant.summary[3:4, ]
cat("\nInterquartile Range:")
lapply(byPregnant.df, IQR, na.rm = T)
```

```r
for(value in c("mean", "median", "IQR")){
    cat(paste0("\n\n", value, ":\n"))
    for(index in seq_along(byPregnant.df)){
        cat(paste0("\n\t", names(byPregnant.df)[index]), "\n\n")
        print(tapply(byPregnant.df[[index]], diabetes.df$pregnant, get(value), na.rm = TRUE))
    }
}
```
