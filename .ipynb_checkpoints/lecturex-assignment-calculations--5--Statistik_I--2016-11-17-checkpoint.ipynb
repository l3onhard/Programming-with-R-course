```r
print_vector <- function(vector){
    cat(sprintf("%s = (%s)\n", 
                as.character(deparse(substitute(vector))), 
                paste(as.character(vector), collapse=", ")))
}
```

```r
# Aufgabe 16
## a)
x = c(4515, 10, 1, 1, 211, 116, 136, 672, 135, 827, 23, 108, 381, 60, 146, 5153, 1243, -6, 473, -3)


g_xANDw_x <- function(x){
    n <- length(x)
    sum_x <- sum(x)
    sum_x_ttpo2 <- sum(x^2)
    x_meanA <- sum(x) / n
    
    s_x <- sqrt(sum((x)^2) / n - x_meanA^2)
    
    sum_for_g_x <- sum(((x - x_meanA) / s_x)^3)
    g_x <- sum_for_g_x / n
    
    sum_for_w_x <- sum(((x - x_meanA) / s_x)^4)
    w_x <- sum_for_w_x / n
    
    for (variableName in c("n", "sum_x", "sum_x_ttpo2", "x_meanA", "s_x", "sum_for_g_x", "g_x", "sum_for_w_x", "w_x")){
        cat(
            sprintf(
                "%s = %s\n", variableName, paste(as.character(eval(parse(text = variableName))), collapse = ", ")
            )
        )
    }
}

g_xANDw_x(x)
```

```r
## b)
bodyHeight <- c(160, 160, 164, 164, 174, 176, 176, 177, 184, 186, 187, 187, 
                166, 170, 179, 179, 188, 194, 170, 170, 174, 180, 180, 182, 196, 197, 200)
table(cut(bodyHeight, breaks = 8))
g_xANDw_x(x = bodyHeight)
hist(bodyHeight)
```
